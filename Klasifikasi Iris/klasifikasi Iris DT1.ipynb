```python
import pandas as pd
import numpy as np
from sklearn import tree
```

```python
from google.colab import drive #memanggil file dengan gdrive
drive.mount('/content/drive')
```

```python
Iris = pd.read_csv('/content/drive/MyDrive/BK/Latihan SC/Iris.csv')
Iris.head()
```

```python
Iris["Species"] = pd.factorize(Iris.Species)[0] #mengubah kolom "species" dari string menjadi unique int
Iris = Iris.drop(labels="Id", axis=1) 
Iris = Iris.to_numpy() 
dataTraining = np.concatenate((Iris[0:40, :], Iris[50:90, :]), axis=0) 
dataTesting = np.concatenate((Iris[40:50, :], Iris[90:100, :]), axis=0) 
```

```python
print(dataTesting)
len(dataTesting)
```

```python
inputTraining = dataTraining[:, 0:4]
inputTesting = dataTesting[:, 0:4]
labelTraining = dataTraining[:,4]
labelTesting = dataTesting[:,4]
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model = model.fit(inputTraining, labelTraining)
```

```python
hasilPrediksi = model.predict(inputTesting)
print("Label sebenarnya : ", labelTesting)
print("Hasil Prediksi : ", hasilPrediksi)
```

```python
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi benar : ", prediksiBenar, "data")
print("Prediksi salah : ", prediksiSalah, "data")
print("Akurasi : ", prediksiBenar/(prediksiBenar+prediksiSalah)
      *100, "%")
```
